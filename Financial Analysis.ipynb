```python
import re
def clean_currency(val):
     val = str(val)
     val = re.sub(r"[^\d.]", "", val)
     return float(val) if val else 0.0

df["Revenue"] = df["Revenue"].apply(clean_currency)
df["Expenses"] = df["Expenses"].apply(clean_currency)
```

```python
conn = sqlite3.connect("/content/sales_data.db")
df.to_sql("sales", conn, if_exists="replace", index=False )
```

```python
query = """
SELECT
    Region,
    SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY Region
ORDER BY Total_Revenue DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Revenue) AS Total_Revenue
FROM sales;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM (Expenses) AS Total_Expenses
FROM sales;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Revenue) - SUM(Expenses) AS Total_Profit
FROM sales;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    strftime('%Y-%m', Date) AS Month,
    SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
     strftime('%Y-%m', Date) AS Month,
     SUM(Revenue) - SUM(Expenses) AS Total_Profit
FROM sales
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Region,
    SUM(Revenue) - SUM(Expenses) AS Total_Profit
FROM sales
GROUP BY Region
ORDER BY Total_Profit DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Product,
    SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY Product
ORDER BY Total_Revenue DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Product,
    SUM(Quantity) AS Units_Sold
FROM sales
GROUP BY Product
ORDER BY Units_Sold DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
     AVG(Revenue) AS Avg_Revenue_Per_Order
FROM sales;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM sales
ORDER BY Revenue DESC
LIMIT 1;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM sales
ORDER BY Revenue ASC
LIMIT 1;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Region,
    COUNT(*) AS Number_of_Orders
FROM sales
GROUP BY Region
ORDER BY Number_of_Orders DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Product,
    AVG((Revenue-Expenses) * 1.0 / Revenue) * 100 AS Avg_Profit_Margin_Percent
FROM sales
GROUP BY Product
ORDER BY Avg_Profit_Margin_Percent DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
     strftime('%Y', Date) AS Year,
     ((CAST(strftime('%m', Date) AS INTEGER) - 1) / 3) + 1 AS Quarter,
     SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY Year, Quarter
ORDER BY Year, Quarter;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    strftime('%Y-%m', Date) AS Month,
    SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY Month
ORDER BY Total_Revenue DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Product,
    SUM(Revenue) AS Total_Revenue
FROM sales
GROUP BY Product
HAVING SUM(Revenue) > 100000
ORDER BY Total_Revenue DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Region,
    SUM(Revenue) - SUM(Expenses) AS Total_Profit
FROM sales
GROUP BY Region
ORDER BY Total_Profit DESC
LIMIT 1;
"""
pd.read_sql(query, conn)
```
